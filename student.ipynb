```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('stud (1).csv')
df.head(2)
```

```python
df.info()
```

```python

df['total_score']=(df['math_score']+df['reading_score']+df['writing_score'])
df['average']=df['total_score']/3 
```

```python

fig,axis=plt.subplots(1,3,figsize=(15,7))
plt.subplot(131)
sns.histplot(data=df,x=df.average,bins=30,kde=True, hue='lunch', multiple='stack')
plt.subplot(132)
sns.histplot(data=df,x=df.average,bins=30,kde=True,hue='gender' , multiple= 'stack')
plt.subplot(133)
sns.histplot(data=df,x=df.average,bins=30,kde=True,hue=df.parental_level_of_education, multiple= 'stack')
```
